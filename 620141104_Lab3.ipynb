```python
from bs4 import BeautifulSoup
url2 = 'https://books.toscrape.com/catalogue/category/books/mystery_3/page-1.html'
url3 = 'https://books.toscrape.com/catalogue/category/books/mystery_3/page-2.html'
response2 = requests.get(url2)
response3 = requests.get(url3)
soup = BeautifulSoup(response2.text, 'html.parser')
soup2 = BeautifulSoup(response3.text, 'html.parser')
```

```python
books = soup.find_all('article')
books2 = soup2.find_all('article')
```

```python
ratings= []
titles= [a.find('h3').find('a').string for a in books]
    #title of the book, the price and the review score


prices = [a.find('p', attrs={'class': 'price_color'}).string[1:] for a in books]
all_classes = soup.find_all(class_=True)
class_names = [i['class'] for i in all_classes]
for a in class_names:
  if a[0] == 'star-rating':
    ratings.append(a[1])

for b in books2:
  titles.append(b.find('h3').find('a').string)
  prices.append(b.find('p', attrs={'class': 'price_color'}).string[1:])

all_classes2 = soup2.find_all(class_=True)
class_names2 = [i['class'] for i in all_classes2]
for b in class_names2:
  if b[0] == 'star-rating':
    ratings.append(b[1])
```

```python
df3 = pd.DataFrame({'Title':titles, 'Price':prices, 'Rating': ratings})
```

```python
df3

```

```python
df3.to_csv('Mystery Books.csv')
```
